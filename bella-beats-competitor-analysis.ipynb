```python
# Converting date and time to date_time objects,
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'])
sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay'])

# normalising the sleep_day to strip the time
sleep_day['SleepDay'] = sleep_day['SleepDay'].dt.normalize()
daily_activity['ActivityDate'] = daily_activity['ActivityDate'].dt.normalize()
```

# **Merging the Data sets**


```python
#Renaming the sleepday date 
sleep_day.rename(columns={"SleepDay": "ActivityDate"},inplace= True)
#Merging the data_sets on the dates using a left join
combined_data = pd.merge(daily_activity, sleep_day, on=['Id', 'ActivityDate'], how='left')
combined_data.head()
```

# **Analysis**

```python
# This counts how many rows have activity but NO sleep data
missing_sleep = combined_data['TotalMinutesAsleep'].isnull().sum()
print(f"There are {missing_sleep} days where users tracked activity but not sleep.")
```

```python
#  new dataframe that ONLY contains days where sleep was tracked
sleep_tracked_only = combined_data.dropna(subset=['TotalMinutesAsleep'])

# calculating the average on this filtered dataframe
avg_awake = (sleep_tracked_only['TotalTimeInBed'] - sleep_tracked_only['TotalMinutesAsleep']).mean()

print(f"On days when sleep was tracked, users spent an average of {avg_awake:.2f} minutes awake in bed.")
```

```python
# Calculate average sedentary hours per day
avg_sedentary_h = sleep_tracked_only['SedentaryMinutes'].mean() / 60
print(f"On average, users are sedentary for {avg_sedentary_h:.2f} hours a day.")
```

```python
# Create a Day of Week column
sleep_tracked_only['DayOfWeek'] = pd.to_datetime(sleep_tracked_only['ActivityDate']).dt.day_name()

# Check average steps by day
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
print(sleep_tracked_only.groupby('DayOfWeek')['TotalSteps'].mean().reindex(day_order))
```

```python
# Correlation between Steps and Calories
correlation = sleep_tracked_only['TotalSteps'].corr(sleep_tracked_only['Calories'])
print(f"Correlation between Steps and Calories: {correlation:.2f}")
```

```python
# Calculate Sleep Quality Ratio
sleep_tracked_only['Sleep_Quality_Ratio'] = (sleep_tracked_only['TotalMinutesAsleep'] / sleep_tracked_only['TotalTimeInBed']) * 100

# Average Sleep Quality
avg_quality = sleep_tracked_only['Sleep_Quality_Ratio'].mean()
print(f"Average Sleep Quality: {avg_quality:.2f}%")
```

# **Exporting Summary Data**

```python
df_for_export = sleep_tracked_only.copy()

# Group by ID and calculate averages
final_summary = df_for_export.groupby('Id').agg({
    'TotalSteps': 'mean',
    'Calories': 'mean',
    'TotalMinutesAsleep': 'mean',
    'TotalTimeInBed': 'mean',
    'SedentaryMinutes': 'mean',
    'Sleep_Quality_Ratio': 'mean'
}).reset_index()

final_summary = final_summary.round(2)

# EXPORT
final_summary.to_csv('bellabeat_final_analysis.csv', index=False)

print("File successfully saved to /kaggle/working/bellabeat_final_analysis.csv")
```

# **Visualisations**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='TotalSteps', y='Calories', data=sleep_tracked_only, 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title('Daily Steps vs. Calories Burned')
plt.xlabel('Total Steps')
plt.ylabel('Calories')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Calculate averages for the whole group
avg_very_active = sleep_tracked_only['VeryActiveMinutes'].mean()
avg_fairly_active = sleep_tracked_only['FairlyActiveMinutes'].mean()
avg_lightly_active = sleep_tracked_only['LightlyActiveMinutes'].mean()
avg_sedentary = sleep_tracked_only['SedentaryMinutes'].mean()

# Data for Pie Chart
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
sizes = [avg_very_active, avg_fairly_active, avg_lightly_active, avg_sedentary]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Average Daily Activity Breakdown')
plt.show()
```

```python
# Ensure DayOfWeek is ordered correctly
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sleep_tracked_only['DayOfWeek'] = pd.Categorical(sleep_tracked_only['DayOfWeek'], categories=day_order, ordered=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='DayOfWeek', y='TotalSteps', data=sleep_tracked_only, palette='viridis', errorbar=None)

plt.title('Average Steps by Day of the Week')
plt.xlabel('Day of the week')
plt.ylabel('Average Steps')
plt.axhline(sleep_tracked_only['TotalSteps'].mean(), color='red', linestyle='--', label='Overall Average')
plt.legend()
plt.show()
```
